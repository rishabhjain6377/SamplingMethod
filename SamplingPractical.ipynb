# Random Sampling or Probabilistics Sampling

 Type of sampling in which each unit of population has some chances or probability of selection in sample.

# 1.Simple Random Sampling

A type of random sampling in which each unit of population has equal chances or probablity of selection.

**NOTE**: In random sampling there is some probability but not equal.

**Steps involved in simple random sampling**

    1. A unique random number is assigned to each unit or observation. This makes it 
    completely unbaised duuring selection.
    
    2.There are many techniques used to select units after assigning random number:
        a. Lottery Method
        
        b. Random number generator software(like python)
        
        c. Random number list
       
**Advantages**

    probaility of selecting same unit is become zero. 

Let's understand it with an example. We will make a sample of 80 units from **IRIS** dataset.

```python
#List of all 80 unique index.
random_index=[]
i=0
while(i<80):
    rand_num=np.random.randint(1,150) #Already imported above
    if rand_num not in random_index: #so that no index being repeated
        random_index.append(rand_num)
        i=i+1
    else:
        i=i #if same index occurs the set i to current i.
```

To check whether all index are unique we can use set fuction. because set function only gives us unique value.

```python
len(set(random_index))
```

Now let's make a sample

```python
sample=iris.loc[random_index,:] #loc[row_index,column_index] here ":" mean all columns
```
